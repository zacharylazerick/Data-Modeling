```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
Shark_Tank = pd.read_csv('Shark_Tank_US.csv', encoding = "utf-8", encoding_errors = "ignore")
```

```python
## Drop Unneeded Columns
mylist = ['Season Start', 'Season End', 'Pitch Number', 'Startup Name', 'Industry', 
          'Business Description', 'Pitchers City', 'Pitchers Average Age', 'Entrepreneur Names', 
          'Company Website', 'Multiple Entrepreneurs', 'US Viewership', 'Royalty Deal', 'Loan', 
          'Barbara Corcoran Present', 'Mark Cuban Present', 'Lori Greiner Present', 'Robert Herjavec Present', 
          'Daymond John Present', 'Kevin O Leary Present', 'Kevin Harrington Present', 'Guest Name', 'Notes']
Shark_Tank_lite = Shark_Tank.drop(columns = mylist)
```

```python
## Create Subsets of DataFrame for Individual Seasons
## Data Frame Indexing Starts at Row 2 in .csv file

## Season 1
Shark_Tank_S1 = Shark_Tank_lite[0:64]

## Season 4
Shark_Tank_S4 = Shark_Tank_lite[160:263]

## Season 7
Shark_Tank_S7 = Shark_Tank_lite[495:611]

## Season 11
Shark_Tank_S11 = Shark_Tank_lite[895:991]
```

### Season 1 Analysis

```python
## Quick Summary of the Data
v_List = ['Episode Number', 'Got Deal', 'Total Deal Amount', 'Total Deal Equity', 'Barbara Corcoran Investment Amount', 
          'Barbara Corcoran Investment Equity', 'Mark Cuban Investment Amount', 'Mark Cuban Investment Equity', 
          'Lori Greiner Investment Amount', 'Lori Greiner Investment Equity', 'Robert Herjavec Investment Amount', 
          'Robert Herjavec Investment Equity', 'Daymond John Investment Amount', 'Daymond John Investment Equity', 
          'Kevin O Leary Investment Amount', 'Kevin O Leary Investment Equity', 'Guest Investment Amount', 
          'Guest Investment Equity']
Shark_Tank_S1[v_List].describe()
```

A quick look at the data for Season 1 shows that of the 64 entrepuneours, only 27 got deals. This means that 42.1875\% of the show's contestants were successful. This season, the sharks gave out a total of $\$4,890,000$ dollars, averaging out to about $\$181,111$ dollars per successful deal. Now, how much money each shark invest? Daymond John invested $\$1,397,500$ across 9 companies, Barbara Corcoran invested $\$1,282,500$ dollars across 14 companies, Robert Herjavec invested $\$910,000$ across 8 companies, Kevin O'Leary invested $\$840,000$ across 6 companies, Guests invested $\$410,000$ across 4 companies, Mark Cuban invested $\$50000$ dollars on 1 company and Lori Greiner did not invest in any company.

```python
## Data Grouped by Episode
Shark_Tank_S1[v_List].groupby('Episode Number').sum()
```

The table above is a brief summary of how each Shark invested their money each epsiode, along with the total amount of money invested per episode. Here we see that, on average, $1.929$ deals made per episode, costing the sharks a whopping $\$349285.71$ per episode. As mentioned before, Barabra Corcoran was the most active shark this season, successfully making a deal in 11 of the 14 episodes. This season, it appears that Mark Cuban and Lori Greiner were mostly along for the ride. 

```python
## Comparison of Entrepenaurs, which ones got a deal which didn't? Gender/Location, etc.
v_List = ['Episode Number', 'Pitchers Gender', 'Pitchers State', 'Original Ask Amount', 'Original Offered Equity', 
          'Valuation Requested', 'Got Deal', 'Total Deal Amount', 'Total Deal Equity']
Shark_Tank_S1[v_List].drop(columns = ['Episode Number']).groupby('Pitchers Gender').sum()
```

Accounting for the Pitcher's Gender now, of the 27 total deals made, 15 of them were from Pitchers that were Male, 10 from Female, and 2 from Mixed Gender Teams. For those that got deals, Males requested about $\$13557$ per percent equity in their companies, Females were about $\$7444$ per percent equity, and Mixed Teams about $\$8269$ per percent equity. 

```python
Shark_Tank_S1[v_List].drop(columns = ['Episode Number']).groupby('Pitchers State').sum().sort_values('Got Deal', ascending = False)
```

Now, for the Pitcher's State (grouped by Region). Of the 27 deals made, 7 (26\%) were from the Pacific Coast, 2 (7.4\%) from the Southwest, 0 from the Rocky Mountain Region, 5 (18.5\%) from the Midwest, 8 (29.6\%) from the Southeast, and 5 (18.5\%) from the Northeast. An ANOVA test to compare the mean proportion of deals made by region would likely reject the null hypothesis, suggesting a relationship between the Pitcher's Region and whether they got a deal or not.
