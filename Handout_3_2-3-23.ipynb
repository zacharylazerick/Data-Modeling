Here we use a similar construction to generate logistic sequences. We'll also plot the
sequences to explore there behavior.

Here is the logistic function:
    
The recursive formula for the logistic sequence $x_n$ is

    $x_{n+1} = 4x_n(1-x_n)$

$r$ is a parameter. $x_1 =x$ is an initial value for the sequence. 

In the following function definition,limit is the number of points we wish to generate. The parameter
$r$ and initial value $x$ are hard coded into the function. Notice that the keyword *yield* indicates
that the current value of $x$ will be returned.



```python
def logistic(x, r, limit):
     
    # Initialize Variables
    n = 0
    
    # return the value x with yield
    # then compute the next value
    while n < limit:
        yield x
        x = r*x*(1-x)
        n = n + 1
```

Let's generate the first values with $x=0.7$ and $r =1.2$

```python
lvalues1 = []

for i in logistic(.7, 1.2, 25):
    lvalues1.append(i)
    
lvalues1
```

We see that the sequennce appears to converge.

We can also plot this. (The following cell only needs to be run once.)

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc("figure", figsize=(10, 6))
np.set_printoptions(precision=4, suppress=True)
```

Here's the plot:

```python
plt.figure()
plt.plot(lvalues1)
```

By changing $r$ and the inital value $x$ we will see different behaviors.
