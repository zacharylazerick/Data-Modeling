## 1 *first_wordle*

The goal is to create a function wordle that mimics the game wordle. 

The input for the first function *first_wordle* consists of two strings:

    * a five letter word *target*
    * a five letter word *guess*
    
Calling the function *first_wordle(target,guess)* returns 

    * the letters of guess which are the same as the letters 
    of target in the same position, and 
    * the letters of guess which are in target, but not in the 
    correct position
    * And a string of the form '_ _ s _ e ' showing the letters in the correct position.
    
In designing your function, 

    * convert strings to lists and/or sets
    * use set and/or set methods to determine 
        * which letters of the guess are letters of the target, then
        * which of these guess letters are in the correct position and
        * which of these guess letters are not in the correct position
    * in the process you may want to use loops and conditionals
    * once the letters have been determined use print statements to report
      the results
    

```python
def first_wordle(target, guess):
    output = ["_", "_", "_", "_", "_"]; letters_not_in_spot = []
    mytarget = list(target); myguess = list(guess)
    for i in range(len(mytarget)):
        for j in range(len(myguess)):
            if ((i == j) and (mytarget[i] == myguess[j])):
                output.pop(i)
                output.insert(i, mytarget[i])
            elif ((i != j) and (mytarget[i] == myguess[j])):
                if (mytarget[i] not in letters_not_in_spot):
                    letters_not_in_spot.append(mytarget[i])
    
    print("The word is", output)
    if (len(letters_not_in_spot) >= 1):
        print("The letters not in the correct spot are", letters_not_in_spot)

first_wordle('worth', 'throw')
```

```python
## Part C, Improved Code

## Import Random package to generate a Random Integer to set as Wordle Word
import random

def real_deal_wordle():
    output = ["_", "_", "_", "_", "_"]; letters_not_used = [] ## Updated: Create a list of unused letters
    target = wordlist[random.randrange(5757)]; mytarget = list(target); counter = 1
    
    while (counter <= 6):
        myguess = input("What is your five letter word guess? ")
        
        ## Updated: Check to see if the guess is valid. (i.e. has five letters and in the wordlist)
        while (myguess not in wordlist):
            print(myguess, "is not a valid word. Please input a valid word.")
            myguess = input("What is your five letter word guess? ")
            
        letters_not_in_spot = []; counter = counter + 1
        
        for i in range(len(mytarget)):
            for j in range(len(myguess)):
                if ((i == j) and (mytarget[i] == myguess[j])):
                    output.pop(i)
                    output.insert(i, mytarget[i])
                elif ((i != j) and (mytarget[i] == myguess[j])):
                    if (mytarget[i] not in letters_not_in_spot):
                        letters_not_in_spot.append(mytarget[i])
        
        ## Updated: Added a check for unused letters so that the player can keep track of them
        for k in range(len(myguess)):
            if (myguess[k] not in output) and (myguess[k] not in letters_not_in_spot):
                if (myguess[k] not in letters_not_used):
                    letters_not_used.extend(myguess[k])
                    letters_not_used.sort()
        
        print("The letters in the correct spot are", output)
        
        ## Updated: Changed the break statement to a return so that the function ends if the conditional is true
        if (output == mytarget):
            return(print("You have guessed the word in", counter - 1, "tries."))
    
        if (len(letters_not_in_spot) >= 1):
            print("The letters not in the correct spot are", letters_not_in_spot)
            
        ## Updated: Added a conditional to print the list of unused letters
        if (len(letters_not_used) >= 1):
            print("The letters not used are", letters_not_used)
            
    
    print("You failed to guess the word. The word was", target)
```
